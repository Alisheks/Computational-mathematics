**Инфа из семинара**

__Дихотомия__

Пусть имеем не слишком осциллирующую непрерывную функцию $f:[a, b] \rightarrow \mathbb{R}$. Алгоритм дихотомии выглядит следующим образом:

>0. Возьмём $[a, b]$ за начальную локализацию минимума. Фиксируем малый $\varepsilon > 0$.
> </br>
> </br>
>1. Пусть текущая локализация минимума - отрезок $[a_n, b_n]$. Берём точку $c_n = \frac{a_n + b_n}{2}$ и смотрим, в какую сторону от неё функция уменьшается. Иными словами, считаем $f_\text{left} = f(c_n - \varepsilon)$ и $f_\text{right} = f(c_n + \varepsilon)$. 
> </br>
> </br>
>2. Если $f_\text{left} < f_\text{right}$, то локальный минимум слева и $[a_{n+1}, b_{n+1}] = [a_n, c_n]$. Иначе, локальный минимум справа и $[a_{n+1}, b_{n+1}] = [c_n, b_n]$.
> </br>
> </br>
>3. Если длина $[a_{n+1}, b_{n+1}]$ меньше $\varepsilon$, останавливаемся и выдаём среднюю точку. Иначе, на шаг 1.

__Метод чисел Фиббоначи (золотого сечения)__

Если операция вычисления непрерывной функции $f:[a, b] \rightarrow \mathbb{R}$ является __очень дорогой__ по времени, разумно использовать следующий алгоритм __чисел Фиббоначи__, использующий вызовы функции с предыдущих итераций. В нём мы заранее фиксируем $N$ - количество вызовов функции $f$.

В алгоритме используются $F_n$ - стандартные числа Фиббоначи $(1, 1, 2, 3, 5, ...)$.

>1. Задаются начальные границы отрезка $a, b$ и число итераций $N$. Счётчик $n = N$.
>    - Рассчитывают начальные точки деления: $x_1=a+(b-a) \frac{F_{n-2}}{F_n}, \quad x_2=a+(b-a) \frac{F_{n-1}}{F_n}$ 
>    - И значения в них целевой функции: $f_1=f\left(x_1\right), f_2=f\left(x_2\right)$.
></br>
></br>
>2. Уменьшаем счётчик $n = N - 1$
>    - Если $f_1>f_2$, то сужаем отрезок слева $a=x_1, \; x_1=x_2$. И модифицированный шаг 1: $x_2=b-\left(x_1-a\right), \; f_1=f_2, \; f_2=f\left(x_2\right)$.
>    -Иначе, сужаем отрезок справа $b=x_2, \; x_2=x_1$. Модифицированный шаг 1: $x_1=a+\left(b-x_2\right), \; f_2=f_1, \; >f_1=f\left(x_1\right)$.
></br>
></br>
>3. Если $n=1$, то останавливаемся и выдаём $\frac{x_1 + x_2}{2}$.
>    - Иначе, на шаг 2.


Ошибку в вычислении минимума от числа итераций $N$ можно оценить как

$$
\varepsilon_N \approx \cfrac{(\sqrt{5}-1) / 2)^N(b-a)}{2}
$$

Несложно обобщить данный метод на произвольной число итераций с учётом $\Phi=\lim _{n \rightarrow \infty} \frac{F_{n+1}}{F_n}$ (это соответствует $N = \infty$ в алгоритме). В таком случае критерием остановки будет $x_2 - x_1 < \varepsilon$. Такой метод называется __методом золотого сечения__.

```python
def dichotomy_method(func, a, b, epsilon=1e-5):
    while (b - a) > epsilon:
        c = (a + b) / 2
        x1 = c - epsilon / 3
        x2 = c + epsilon / 3

        if func(x1) < func(x2):
            b = x2
        else:
            a = x1

    return (a + b) / 2

def fibonacci_method(func, a, b, iterations=50):
    fib_sequence = [1, 1]

    for _ in range(2, iterations):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])

    rho = (fib_sequence[-3] / fib_sequence[-1]) * (b - a)
    x1 = b - rho
    x2 = a + rho

    for _ in range(iterations - 1):
        if func(x1) < func(x2):
            b = x2
            x2 = x1
            x1 = a + b - x2
        else:
            a = x1
            x1 = x2
            x2 = b - x1

    return (a + b) / 2
```

```python
import numpy as np

def func_a(x):
    return 2 * x**2 - np.log(x)

a, b = 0.1, 2.0
min_dichotomy_a = dichotomy_method(func_a, a, b)
min_fibonacci_a = fibonacci_method(func_a, a, b)

print("Минимум (дихотомия):", min_dichotomy_a)
print("Минимум (Фиббоначи):", min_fibonacci_a)
```

### 2.3. Функция Растригина.

$$
f\left(x_1, x_2\right)=20+x_1^2+x_2^2-10 \cos \left(2 \pi x_1\right)-10 \cos \left(2 \pi x_2\right)
$$

Постройте множество локальных минимумов и максимумов на плоскости используя `minimize`. Минимумы должны быть синими, максимумы - красными. В 3D график функции выглядит примерно так:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def rastrigin(x): #определение функции Растригина
    return 20 + x[0]**2 + x[1]**2 - 10 * (np.cos(2 * np.pi * x[0]) + np.cos(2 * np.pi * x[1]))

#задание диапазонов для построения графика
x_min, x_max = -5.12, 5.12
y_min, y_max = -5.12, 5.12

#создание сетки значений для функции Растригина
x_vals = np.linspace(x_min, x_max, 100)
y_vals = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x_vals, y_vals) #создает координатные матрицы на основе входных векторов
Z = rastrigin([X, Y])

#3D-график функции Растригина
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Функция Растригина')

#нач. догадки для минимизации
initial_guesses = [(-4, -4), (-2, 2), (0, 0), (2, -2), (4, 4)]

#поиск минимума функции Растригина с использованием BFGS
for guess in initial_guesses:
    result = minimize(rastrigin, guess, method='BFGS')
    x_min, y_min = result.x
    z_min = rastrigin(result.x)
    if result.success: #визуализация найденного минимума
        ax.scatter(x_min, y_min, z_min, color='blue', marker='o', s=100)
    else:
        print(f"Не удалось найти минимум для начального значения {guess}")

plt.show()
```

__Инфа из семинара__

**Градиентный спуск**
В общем случае, он выглядит вот так:

>1. Задают начальное приближение и точность расчёта $\vec{x}_{0}, \varepsilon$
></br>
></br>
>2. Рассчитывают $\vec{x}_{n+1}=\vec{x}_n-\alpha \cdot \nabla f\left(\vec{x}_n\right)$, где $\alpha$ - __шаг градиентного спуска__ или __learning rate__ .
></br>
></br>
>3. Проверяют условия остановки:
>    - Если $\left|\vec{x}_{n+1}-\vec{x}_n\right|>\varepsilon,\;\; \left|f\left(\vec{x}_{n+1}\right)-f\left(\vec{x}_n\right)\right|>\varepsilon$ или $\left\|\nabla f\left(\vec{x}_{n+1}\right)\right\|>\varepsilon$ (выбирают одно из условий), то $n=n+1$ и переход к шагу 2.
>    - Иначе $\vec{x}=\vec{x}_n$ и остановка.
></br>

Алгоритм __метода наискорейшего спуска__:

>1. Задают начальное приближение и точность расчёта $\vec{x}_{0}, \varepsilon$
></br>
></br>
>2. Рассчитывают $\vec{x}_{n+1}=\vec{x}_n-\alpha_n \cdot \nabla f\left(\vec{x}_n\right)$, где $\alpha_n$ считают из однопараметрической минимизации:
></br>
></br>
>$$
\alpha_n \stackrel{\text { def }}{=}\underset{\alpha}{\operatorname{argmin}} \; f\left(\vec{x}_n-\alpha \cdot \nabla f\left(\vec{x}_n\right)\right)
$$
></br>
>3. Проверяют условия остановки:
>    - Если $\left|\vec{x}_{n+1}-\vec{x}_n\right|>\varepsilon,\;\; \left|f\left(\vec{x}_{n+1}\right)-f\left(\vec{x}_n\right)\right|>\varepsilon$ или $\left\|\nabla f\left(\vec{x}_{n+1}\right)\right\|>\varepsilon$ (выбирают одно из условий), то $n=n+1$ и переход к шагу 2.
>    - Иначе $\vec{x}=\vec{x}_n$ и остановка.
></br>

На первом рисунке метод наискорейшего спуска, на втором - метод сопряженных градиентов. Это понятно из направлений движения: на первом рисунке мы движемся строго ортогонально линиям уровня, на втором - уже нет, что и характерно для метода СГ.
```python
import numpy as np
from scipy.optimize import minimize

def staib_tang(x): #ф-ия Стайб-Танга - её минимизируем
    n = len(x)
    return 0.5 * np.sum(x**4 - 16 * x**2 + 5 * x)

def staib_tang_gradient(x): #возвращает градиент ф-ии Стайб-Танга
    n = len(x)
    return np.array([2 * x_i * (x_i**2 - 8 * x_i + 5) for x_i in x])

def minimize_func(alpha, f, grad_f, x, d): #для нахождения опт. знач. шага alpha методом оптимизации по одной переменной
    return f(x + alpha * d)

def conjugate_gradient_method(f, grad_f, x0, method_type, max_iter=1000, tol=1e-6): #метод сопр. градиентов
    x = x0 #x0 - нач. прибл.
    grad = grad_f(x)
    d = -grad
    k = 0

    #
    if method_type == 1:
        beta_type = "Hestenes-Stiefel"
        beta_func = lambda new_grad, grad: np.dot(new_grad, new_grad - grad) / np.dot(new_grad - grad, d)
    elif method_type == 2:
        beta_type = "Fletcher-Reeves"
        beta_func = lambda new_grad, grad: np.dot(new_grad, new_grad) / np.dot(grad, grad)
    elif method_type == 3:
        beta_type = "Polak-Ribiere"
        beta_func = lambda new_grad, grad: np.dot(new_grad, new_grad - grad) / np.dot(grad, grad)
    elif method_type == 4:
        beta_type = "Fletcher"
        beta_func = lambda new_grad, grad: np.dot(new_grad, new_grad) / np.dot(grad, d)
    elif method_type == 5:
        beta_type = "Liu-Storey"
        beta_func = lambda new_grad, grad: max(0, np.dot(new_grad, new_grad - grad) / np.dot(grad, d))
    elif method_type == 6:
        beta_type = "Dai-Yuan"
        beta_func = lambda new_grad, grad: max(0, np.dot(new_grad, new_grad - grad) / np.dot(new_grad, new_grad))
    elif method_type == 7:
        beta_type = "Hager-Zhang"
        beta_func = lambda new_grad, grad: max(0, np.dot(new_grad, new_grad - grad) / np.dot(grad, new_grad - grad))

    #цикл - до момента, когда норма градиента становится меньше порога или достигнуто максимальное число итераций).
    while np.linalg.norm(grad) > tol and k < max_iter:
        alpha = minimize(minimize_func, 0, args=(f, grad_f, x, d), bounds=[(0, None)]).x
        x = x + alpha * d
        new_grad = grad_f(x)
        beta = beta_func(new_grad, grad)
        d = -new_grad + beta * d
        grad = new_grad
        k += 1

    return x, k, beta_type

initial_guesses = [(0.1, -1), (0.1, -1.1), (0, -1)]

for guess in initial_guesses:
    for method_type in range(1, 8):
        result, iterations, beta_type = conjugate_gradient_method(staib_tang, staib_tang_gradient, np.array(guess), method_type)
        print(f"Начальное приближение: {guess}, Результат: {result}, Количество итераций: {iterations}, Метод: {beta_type}")
```
