### 1.1. Апостериорное распределение для параметра распределения Пуассона.


Количество срабатываний счетчика Гейгера за минуту $n$ подчиняется распределению Пуассона:
$$
P_\lambda(n)=\frac{\lambda^n}{n !} e^{-\lambda} \text {. }
$$
1. В ходе эксперимента счетчик Гейгера сработал за минуту $m$ раз. С помощью теоремы Байеса определите апостериорное распределение на $\lambda$. Указание: априорную плотность вероятности $\lambda$ можно считать постоянной (так как мы изначально ничего не знаем про $\lambda$ ) .
2. Эксперимент повторили еще раз, в этот раз счетчик Гейгера сработал за минуту $m^{\prime}$ раз. Как обновилось апостериорное распределение на $\lambda$ ?

**Примечание.**

Такая плотность вероятности не будет нормируема. Чтобы сделать рассуждение более строгим, можно ввести обрезку на очень больших $\lambda$ (так как это нереалистичные значения). Другими словами, можно считать, что априорная плотность вероятности $p_0(\lambda)-$ это какая-то очень медленно меняющаяся функция и как-то убывающая на бесконечности. Тогда в числителе и знаменателе формулы Байеса она будет домножаться на гораздо более быструю функцию и поэтому можно заменить $p_0(\lambda) \rightarrow p_0(0)$. Константа $p_0(0)$ должна сократиться в ходе вычислений.


Пусть $X$ - случайная величина (число срабатываний счетчика Гейгера за минуту). Тогда априорная плотность вероятности $p(\lambda)$ постоянна, и мы можем записать теорему Байеса следующим образом:

1) Первый эксперимент:
\begin{align*}
P(\lambda | X=m) &\propto P(X=m | \lambda) \cdot p(\lambda) \
&\propto \frac{\lambda^m}{m!} e^{-\lambda} \cdot p_0(\lambda),
\end{align*}

где $p_0(\lambda)$ - априорная плотность вероятности. Из примечания заменим $p_0(\lambda)$ на $p_0(0)$ , что позволит нам избавиться от константы при дальнейших вычислениях.

2) Второй эксперимент:
После второго эксперимента с $m'$ срабатываниями, апостериорное распределение обновится

\begin{align*}
P(\lambda | X=m, X'=m') &\propto P(X=m | \lambda) \cdot P(X'=m' | \lambda) \cdot p(\lambda) \
&\propto \frac{\lambda^m}{m!} e^{-\lambda} \cdot \frac{\lambda^{m'}}{m'!} e^{-\lambda} \cdot p_0(\lambda) \
&\propto \frac{\lambda^{m+m'}}{(m+m')!} e^{-2\lambda} \cdot p_0(\lambda).
\end{align*}

Апостериорное распределение на $\lambda$ после второго эксперимента будет иметь функцию плотности вероятности $\frac{\lambda^{m+m'}}{(m+m')!} e^{-2\lambda} \cdot p_0(0)$.

### 1.2. Апостериорное распределение для аргумента нормального распределения.

Пусть имеется априорное распределение на вектор $\boldsymbol{x}$, задаваемое симметричной положительно определенной матрицей $A$ :
$$
p_0(\boldsymbol{x})=\frac{1}{Z} e^{-\frac{\boldsymbol{x}^T A \boldsymbol{x}}{2}} .
$$
Было произведено измерение величин $\boldsymbol{x}$, которое дало значение $\boldsymbol{x}_1$. Найдете апостериорное распределение на $\boldsymbol{x}$.

Hint: Мы изначально ничего не знаем об $A$ - надо ввести на него какое-то априорное распределение. Предлагается брать равномерное.

__Решение:__

>- $P(A)$ - __априорная вероятность__ гипотезы $A$ (до получения данных о истинности события $B$);
>
>- $P(A \mid B)$ - вероятность гипотезы $A$ при наступлении события $B$ (уточнённая, __апостериорная вероятность__);
>
>- $P(B \mid A)$ - вероятность наступления события $B$ при истинности гипотезы $A$ (__правдоподобие, likelihood__;
>
>- $P(B)$ - полная вероятность наступления события $B$.

Для нахождения апостериорного распределения вектора $\boldsymbol{x}$ после измерения $\boldsymbol{x}_1$, мы можем использовать формулу Байеса. Априорное распределение на $\boldsymbol{x}$ задано следующим образом:
$$ p_0(\boldsymbol{x}) = \frac{1}{Z} e^{-\frac{\boldsymbol{x}^T A \boldsymbol{x}}{2}}. $$

Функция правдоподобия измерения $\boldsymbol{x}_1$ в предположении нормального распределения с некоторой матрицей ковариации $B$ выглядит так:
$$ p(\boldsymbol{x}_1 | \boldsymbol{x}) = \frac{1}{Z_1} e^{-\frac{(\boldsymbol{x}_1 - \boldsymbol{x})^T B^{-1} (\boldsymbol{x}_1 - \boldsymbol{x})}{2}}. $$

Теперь, используя формулу Байеса, апостериорное распределение на $\boldsymbol{x}$ после измерения $\boldsymbol{x}_1$ можно записать как пропорциональное произведение априорного распределения и функции правдоподобия:
$$ p(\boldsymbol{x} | \boldsymbol{x}_1) \propto p_0(\boldsymbol{x}) \cdot p(\boldsymbol{x}_1 | \boldsymbol{x}). $$

Подставим значения и упростим выражение:
$$ p(\boldsymbol{x} | \boldsymbol{x}_1) \propto e^{-\frac{\boldsymbol{x}^T A \boldsymbol{x}}{2}} \cdot e^{-\frac{(\boldsymbol{x}_1 - \boldsymbol{x})^T B^{-1} (\boldsymbol{x}_1 - \boldsymbol{x})}{2}}. $$

Объединим экспоненты и упростим:
$$ p(\boldsymbol{x} | \boldsymbol{x}_1) \propto e^{-\frac{1}{2}(\boldsymbol{x}^T A \boldsymbol{x} - \boldsymbol{x}^T A^{-1} \boldsymbol{x}_1 - \boldsymbol{x}_1^T B^{-1} \boldsymbol{x} + \boldsymbol{x}_1^T A^{-1} \boldsymbol{x}_1)}. $$

Таким образом, апостериорное распределение на $\boldsymbol{x}$ после измерения $\boldsymbol{x}_1$ имеет вид: $$ p(\boldsymbol{x} | \boldsymbol{x}_1) \propto e^{-\frac{1}{2}(\boldsymbol{x}^T (A + B^{-1}) \boldsymbol{x} - \boldsymbol{x}^T A^{-1} \boldsymbol{x}_1 - \boldsymbol{x}_1^T B^{-1} \boldsymbol{x} + \boldsymbol{x}_1^T A^{-1} \boldsymbol{x}_1)}. $$

(Нормировочная константа может быть сложной для вычисления, и её может потребоваться аппроксимировать)


### 1.3. L1-регуляризация.

Для стандартного набора данных для задачи регрессии (см. например load_diabetes из sklearn.datasets) продемонстрируйте, как веса обращаются в ноль по мере увеличения коэффициента $\mu$ $L_1$-регуляризации. Разрешается использовать библиотечную реализацию регрессии, как в примере ниже:

__Инфа из семинара:__

Наивно, эвристически, бороться с переобучением можно удаляя этот симптом больших весов (сложности модели). Иными словами, добавим в функцию ошибок слагаемое, отвечающие штрафу за такие скачки.

$$
\mathcal{L} \rightarrow \mathcal{L}+R(\mathbf w)
$$

Такой штраф должен ограничивать количество признаков (чтобы их не стало больше количества точек в train-е) или рост весов модели $\mathbf w$ (чтобы функция не могла быстро осциллировать).

...

- __L1-регуляризация__

Иным выбором функции регуляризации может служить __L1-регуляризация__ (__lasso__). Вместо квадрата модуля, мы возьмём модуль вектора непосредственно.

$$
R(\mathbf w)=\mu\|\mathbf w\|_1=\mu \sum_{\alpha=1}^m\left|\text w_\alpha\right|, \quad \mu > 0
$$

Таким образом, новая функция ошибок будет выглядеть как

$$
\mathcal{L}=\frac{1}{N}\left\|X\mathbf w-Y\right\|^2+\mu \sum_\alpha\left|\text w_\alpha\right|
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import Lasso #L1- lasso

data, target = load_diabetes(return_X_y=True)

# Кф регуляризации (для управления сложностью модели и предотвращения переобучения (иначе сложно обобщить данные))
a = 1e-6 #нач. знач. кф рег.
n = 500 #кол-во знач., равном. распр.
alpha_linspace = np.linspace(a, 1, n) #создает массив значений коэффициента регуляризации.

# Буфер для сохранения весов модели для каждой фичи и каждого значения коэффициента регуляризации.
buf = np.zeros((n, 10))

for i in range(n):
    lasso = Lasso(alpha=alpha_linspace[i]) #инициализация модели Lasso с текущим значением коэффициента регуляризации.
    lasso.fit(data, target) #обучение модели на данных.
    values = lasso.coef_ #получение весов модели.
    buf[i] = values #сохр. в буфер
 
plt.figure(figsize=(20, 20))
for i in range(10):
    plt.subplot(5, 2, i + 1)
    plt.plot(alpha_linspace, buf[:, i], label=f'Feature {i + 1}')
    plt.legend()

plt.show()
```

### 1.5. Лассо Тибширани: связь L1 и L2 между собой.


Покажите, что задача минимизации квадратичной функции потерь с дополнительным ограничением (лассо Тибширани):
$$
\mathcal{L}=\|X w-y\|^2 \rightarrow \min _w, \quad \sum_\alpha\left|w_\alpha\right|<C
$$
эквивалентна L1-регуляризации. Указание: можно воспользоваться условиями Каруша - Куна — Таккера (обобщение метода Лагранжа).
[Link](https://ru.wikipedia.org/wiki/%D0%A3%D1%81%D0%BB%D0%BE%D0%B2%D0%B8%D1%8F_%D0%9A%D0%B0%D1%80%D1%83%D1%88%D0%B0_%E2%80%94_%D0%9A%D1%83%D0%BD%D0%B0_%E2%80%94_%D0%A2%D0%B0%D0%BA%D0%BA%D0%B5%D1%80%D0%B0) .

Задача минимизации квадратичной функции потерь с дополнительным ограничением (lasso Тибширани):
$$
\mathcal{L} = \|Xw-y\|^2 \rightarrow \min_{w}, \quad \sum_{\alpha}\left|w_{\alpha}\right| < C
$$

Эквивалентная L1-регуляризация:
$$
\mathcal{L}_{\text{L1}} = \|Xw-y\|^2 + \lambda \sum_{\alpha}\left|w_{\alpha}\right|
$$

Функция Лагранжа:
$$
L(w, \lambda) = \|Xw-y\|^2 + \lambda \left(\sum_{\alpha}\left|w_{\alpha}\right| - C\right)
$$

Условия Каруша-Куна-Таккера:
$$
\frac{\partial L}{\partial w} = 2X^T(Xw - y) + \lambda \cdot \text{sign}(w) = 0
$$
$$
\frac{\partial L}{\partial \lambda} = \sum_{\alpha}\left|w_{\alpha}\right| - C = 0
$$

Решение первого уравнения:
$$
X^TXw - X^Ty + \lambda \cdot \text{sign}(w) = 0
$$

Исключив \(X^TXw - X^Ty\), получаем:
$$
\nabla||w||_2 + \frac{\lambda}{2} \nabla||w||_1 = 0
$$

Таким образом, мы получаем связь между L2-регуляризацией и L1-регуляризацией в контексте лассо Тибширани.
